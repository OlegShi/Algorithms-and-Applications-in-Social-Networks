```python
import warnings
warnings.filterwarnings('ignore')
import networkx as nx
import matplotlib.pyplot as plt

def PreferentialAttachment(G):
    max_PA = 0
    CG = nx.complement(G);
    maxEdge = list(CG.edges())[0]
    for edge in list(CG.edges()):
        pa = G.degree(edge[0])*G.degree(edge[1])
        if pa > max_PA:
            maxEdge = edge
            max_PA = pa
    print("The edge that has the highest chance to be created in this network is ", maxEdge, "with Preferential Attachment equals to", max_PA, "\n")

 
G=nx.Graph()
G.add_node("A")
G.add_node("B")
G.add_node("C")
G.add_node("D")
G.add_node("E")
G.add_node("F")

G.add_edge("A", "B")
G.add_edge("A", "C")
G.add_edge("A", "D")
G.add_edge("B", "C")
G.add_edge("B", "E")
G.add_edge("D", "E")
G.add_edge("D", "F")
        
nx.draw(G, with_labels=True, node_size=1000, node_color="skyblue")
plt.show()
print("\n")
PreferentialAttachment(G)

    
```
