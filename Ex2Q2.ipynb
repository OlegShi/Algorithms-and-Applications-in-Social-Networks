```python
import matplotlib.pyplot as plt
import networkx as nx

def k_clique_communities_detection(G, k):
    
    cliques = [c for c in nx.find_cliques(G)]
    length = len(cliques)
    Matrix = [[0 for x in range(length)] for y in range(length)]
    G_Matrix=nx.Graph()
    for i in range(0, length):
        G_Matrix.add_node(i)
        
    #Create clique overlap matrix
    for i in range(0, length):
        for j in range(0, length):
            count = 0
            for n1 in cliques[i]:
                for n2 in cliques[j]:
                    if n1==n2:
                        count += 1
            Matrix[i][j] = count 
        
    
    #Threshold matrix with k-1
    for i in range(0, length):
        for j in range(0, i+1):
            if i==j:
                if Matrix[i][j]<k:
                    Matrix[i][j]=0
                else:
                    Matrix[i][j]=1
                    
            else:
                if Matrix[i][j]<(k-1):
                    Matrix[i][j]=0
                else:     
                    Matrix[i][j]=1
                    G_Matrix.add_edge(i, j)
                    
    
    #Communities are connected components
    components = [c for c in nx.connected_components(G_Matrix)]
    num_of_components = len(components)
    communities = []
    for i in range(0, num_of_components):
        com = []
        if len(components[i])>1:
            for c in components[i]:
                for n in cliques[c]:
                    if n in com:
                        pass
                    else:
                        com.append(n)      
            communities.append(com)
        else:
            for c in components[i]:
                if len(cliques[c])==k:
                    communities.append(cliques[c])
                 
    return communities
    
    
    
    
```

```python
import matplotlib.pyplot as plt
import networkx as nx

def buildGraphFromFile(G, input):        
    input_data = open(input, 'rU')
    for line in input_data:
        line = line.strip('\n')
        line = str(line)
        line = line.split(" ")
    
        G.add_edge(int(line[0]),int(line[1]))
            

G=nx.Graph()
buildGraphFromFile(G, "communities.txt")

#biggest connected component
Gc = max(nx.connected_component_subgraphs(G), key=len)

#run our algo on the biggest connected component as stated in class
comm = k_clique_communities_detection(Gc , 4)

comm_id = 0
for c in comm:
    print("Comm id:", comm_id)
    print(c)
    comm_id += 1
```
